```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
X,y = make_blobs(n_samples=200,centers=2,random_state=64)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel("x0")
plt.ylabel("x1")
plt.show()
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
slope = model.coef_
intercept = model.intercept_
```

```python
y_train_pred = -(slope[0][0] * X_train[:,0] + intercept)/slope[0][1] 
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(X_train[:,0],y_train_pred,color="green")
plt.xlabel("x0")
plt.ylabel("x1")
plt.show()
```

```python
print(model.score(X_test,y_test))
```

```python
X,y = make_blobs(centers=4,random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
model = LogisticRegression(multi_class='warn')
model.fit(X_train,y_train)
```

```python
slope = model.coef_
intercept = model.intercept_
color_list = model.classes_
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1],c=y)
for i in range(intercept.shape[0]):
    y_train_pred = -(slope[i][0] * X_train[:,0] + intercept[i])/slope[i][1]
    plt.plot(X_train[:,0],y_train_pred)
plt.xlim(X_train[:,0].min(),X_train[:,0].max())
plt.ylim(X_train[:,1].min(),X_train[:,1].max())
plt.show()
```

```python
model.score(X_test,y_test)
```
